# Exploratory data analysis

```python
# Used to supress the following matplotlib warning:
# matplotlib/__init__.py:872: UserWarning: axes.color_cycle is deprecated and replaced with axes.prop_cycle; please use the latter.
#   warnings.warn(self.msg_depr % (key, alt_key))
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
train_data = pd.read_csv('raw_data/train.csv')
train_data
```

```python
test_data = pd.read_csv('raw_data/test.csv')
test_data
```

## Missing values

```python
100 * train_data.isnull().sum() / train_data.shape[0]
```

```python
100 * test_data.isnull().sum() / test_data.shape[0]
```

* Roughly 75% missing in `Cabin` and 20% missing in `Age`. The rest of the data set is pretty complete. 
* `Cabin` could probably be discarded as a predictor in a first attempt. 
* `Age` might be an important variable -- an imputation strategy more elaborate than `"mean"` may be useful.
* Training samples with missing `Embarked` values could be ignored.
* `Fare` in test sample could be imputed with average `Fare` for each `Pclass`.

Check some variables that might be useful to predict `Age`.

```python
g = sns.lmplot(x='Age', y='Fare', hue='Pclass', size=6, aspect=2, data=train_data)
g.set(xlim=(0, None), ylim=(0, None))
```

```python
sns.barplot(x='Sex', y='Age', data=train_data)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
sns.barplot(x='SibSp', y='Age', order=range(9), data=train_data, ax=ax[0])
sns.barplot(x='Parch', y='Age', order=range(7), data=train_data, ax=ax[1])
```

# Visualize individual variables

```python
sns.barplot(x='Pclass', y='Survived', order=(1, 2, 3), data=train_data)
```

```python
sns.barplot(x='Sex', y='Survived', data=train_data)
```

```python
train_data['AgeGroup'] = pd.cut(train_data['Age'], 5)
sns.barplot(x='AgeGroup', y='Survived', data=train_data)
```

```python
train_data['FareGroup'] = pd.cut(train_data['Fare'], 3)
sns.barplot(x='FareGroup', y='Survived', data=train_data)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
sns.barplot(x='SibSp', y='Survived', order=range(9), data=train_data, ax=ax[0])
sns.barplot(x='Parch', y='Survived', order=range(7), data=train_data, ax=ax[1])
```

Combine `SibSp` and `Parch` into a single variable?

```python
train_data['Relatives'] = train_data['SibSp'] + train_data['Parch']
sns.barplot(x='Relatives', y='Survived', order=range(11), data=train_data)
```
