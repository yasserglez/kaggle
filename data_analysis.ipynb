## Check for missing values

```python
100 * train.isnull().sum() / train.shape[0]
```

```python
100 * test.isnull().sum() / test.shape[0]
```

Roughly 20% missing in `Age`, 77% missing in `Cabin`, and the rest of the data set is pretty complete. `Cabin` could probably be discarded as a predictor in a first attempt. `Age` might be an important variable -- impute missing values? Training samples with missing `Embarked` values could also be ignored.

# Visualize individual variables

```python
sns.barplot(x='Pclass', y='Survived', order=(1, 2, 3), data=train)
```

```python
sns.barplot(x='Sex', y='Survived', data=train)
```

```python
train['AgeGroup'] = pd.cut(train.Age, 5)
sns.barplot(x='AgeGroup', y='Survived', data=train)
```

```python
train['FareGroup'] = pd.cut(train.Fare, 3)
sns.barplot(x='FareGroup', y='Survived', data=train)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
sns.barplot(x='SibSp', y='Survived', order=range(9), data=train, ax=ax[0])
sns.barplot(x='Parch', y='Survived', order=range(7), data=train, ax=ax[1])
```

Consider combining `SibSp` and `Parch` into a single variable?

```python
train['Relatives'] = train.SibSp + train.Parch
sns.barplot(x='Relatives', y='Survived', order=range(11), data=train)
```
